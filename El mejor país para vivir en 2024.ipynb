6. Muestra los indice de felicidad por región

```python
region_feliz = df.groupby('region')['WorldHappiness2022'].mean().sort_values(ascending=False)

region_feliz.plot(kind='bar')
plt.xlabel('Región')
plt.ylabel('Indicie de Felicidad media 2022')
plt.title('Indicie de Felicidad por Región en 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Explicación :

Las regiones están ordenadas de mayor a menor felicidad. Este gráfico es útil para comparar rápidamente la felicidad percibida en diferentes partes del mundo.

7. Muestra los 10 paises más felices del mundo

```python
top_10_felices = df.nlargest(10,'WorldHappiness2022')[['country','region','WorldHappiness2022']]
colors = plt.cm.rainbow(np.linspace(0, 1, len(top_10_felices)))
for i, (country, happiness) in enumerate(zip(top_10_felices['country'], top_10_felices['WorldHappiness2022'])):
    plt.scatter(country, happiness, color=colors[i], label=country, s=100) 
plt.xlabel('País')
plt.ylabel('Indicie de Felicidad 2022')
plt.xticks(rotation=45)
plt.title('Top 10 de los países más felices en 2022')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Explicación :

# El gráfico revela que Finlandia tiene la puntuación más alta de felicidad, seguida por otros países principalmente europeos. Los nombres de los países están claramente etiquetados y la rotación de 45 grados de las etiquetas en el eje X asegura que todos los nombres sean legibles

9. Muestra un mapa de felicidad de todos los países del mundo

```python
import plotly.express as px
import pycountry
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
def get_iso_alpha_3(country):
    try:
        return pycountry.countries.lookup(country).alpha_3
    except:
        return None

df['iso_alpha'] = df['country'].apply(get_iso_alpha_3)

# Personalizamos la escala de colores
color_scale = px.colors.diverging.RdYlGn[::-1]  # Invertimos la escala para que los colores cálidos indiquen alta felicidad

# Crear el mapa coroplético
mapa_felicidad = px.choropleth(df, 
                               locations="iso_alpha", 
                               color="WorldHappiness2022",
                               hover_name="country", 
                               color_continuous_scale=color_scale,
                               title="Niveles de Felicidad Mundial en 2022",
                               labels={'WorldHappiness2022': 'Índice de Felicidad'})

# Mostramos el mapa
mapa_felicidad.show()
```

Explicación :

# Este código utiliza Plotly Express para visualizar cómo se distribuye el índice de felicidad a lo largo de los países del mundo en un mapa interactivo, con un enfoque visual en los niveles de felicidad
